```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Load dataframe
df = pd.read_excel('Absenteeism_at_work.xls')
df.head()
```

```python
data_scaled=normalize(df)
data_scaled=pd.DataFrame(data_scaled,columns=df.columns)
```

```python
data_scaled.head()
```

```python

plt.figure(figsize=(10,7))
dend=dendrogram(linkage(data_scaled,method='ward'),10)
plt.axhline(y=0.005,color='r',linestyle='--')
```

```python
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean', linkage='ward')
```

```python
cluster.fit_predict(data_scaled)
```

```python
plt.figure(figsize=(10,7))
X_column='Reason for absence'
Y_column='Month of absence'
plt.scatter(data_scaled[X_column],data_scaled[Y_column],c=cluster.labels_)
plt.title('Clusters of Employees')
plt.xlabel(X_column)
plt.ylabel(Y_column)
plt.show()
```

```python


# Assigning cluster labels to the original data
df['Cluster'] = cluster.fit_predict(data_scaled)
table = df.groupby(['Cluster']).mean()
table = table.iloc[:, :20]
table.transpose()
```
